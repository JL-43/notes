## Download the model checkpoint

https://huggingface.co/deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"

model = AutoModelForCausalLM.from_pretrained(
    model_name,).to("cuda")
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

## Provide a prompt & generation parameters

```python
prompt = "Create for me a vigenere ciphered text, using the key 'elon', and the text being 'do you ever feel, like a plastic bag drifting through the wind, wanting to start again?'"
messages = [
    {"role": "system", "content": "You are an extremely focused and to the point assistant."},
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)
```

## Generate text

```python
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=2048
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]
```

## Decode response

```python
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

## Voila, enjoy the response!

```python
print(response)
```
